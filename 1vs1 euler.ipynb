Programa para determinar por fuerza bruta el equivalente en potencias de numeros primos de un numero.

```python
import time

def mide_tiempo(funcion):
    def funcion_medida(*args, **kwargs):
        inicio = time.time()
        c = funcion(*args, **kwargs)
        print(time.time() - inicio)
        return c
    return funcion_medida

def prime_factors(n):
    """
    Devuelve una lista con los factores primos de n.
    """
    i = 2
    factors = []
    while i * i <= n:
        if n % i:
            i += 1
        else:
            n //= i
            factors.append(i)
    if n > 1:
        factors.append(n)
    return factors
@mide_tiempo
def prime_factorization(n):
    """
    Descompone n en un producto de potencias de números primos.
    """
    factors = prime_factors(n)
    unique_factors = list(set(factors))
    factorization = []
    for factor in unique_factors:
        power = factors.count(factor)
        factorization.append((factor, power))
    return factorization

# Ejemplo de uso
n = 98212
factorization = prime_factorization(n)
print(f"La factorización de {n} en un producto de potencias de números primos es:")
print(" x ".join([f"{factor}^{power}" for factor, power in factorization]))
```
