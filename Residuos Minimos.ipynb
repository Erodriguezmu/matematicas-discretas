¿Que dia de la semana va a ser dentro de 1000 dias si hoy es jueves?
jueves equivale al 4 dia de la semana
debemos resolver la siguiente congruencia:
```
1000 ≡ ? (mod 7)
142 ≤ 1000/7 < 143
a - qn
1000 - 142(7)=1000-994=6
1000 ≡ 6(mod 7)
```
Como el dia actual es jueves (4) se le agrega el modulo(6) y sabremos que dia sera dentro de los 1000 dias

```
  4         5       6         0        1       2         3
jueves | viernes |sabado | domingo | lunes | martes | miercoles |
```
