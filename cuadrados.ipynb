expresar un n√∫mero como una suma de cuadrados (pero con la menor cantidad de cuadrados).


```python
import math

def sum_of_squares(n):
    """
    Devuelve una lista con la menor cantidad de cuadrados cuya suma es n.
    """
    squares = []
    i = 1
    while i * i <= n:
        squares.append(i * i)
        i += 1
    counts = [float("inf")] * (n + 1)
    counts[0] = 0
    for i in range(1, n + 1):
        for square in squares:
            if i - square >= 0:
                counts[i] = min(counts[i], counts[i - square] + 1)
    result = []
    i = n
    while i > 0:
        for square in squares:
            if i - square >= 0 and counts[i - square] + 1 == counts[i]:
                result.append(square)
                i -= square
                break
    return result

# Ejemplo de uso
n = 500
squares = sum_of_squares(n)
print(f"La menor cantidad de cuadrados que suman {n} es:")
print(" + ".join([f"{int(math.sqrt(square))}^2" for square in squares]))
```
