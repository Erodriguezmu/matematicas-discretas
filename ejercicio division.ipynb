Para cada uno de los siguientes a y n, encuentre el cociente y el resto de dividir a sobre b y escriba la ecuacion a= qn+r

```
1) a=59 n=7   ->  8<59/7<9
r = a-qn = 59-8*(7) = 59-56 = 3
59 = 8*7+3
```
```
2) a=84 n=12   ->  7=84/12
r = a-qn = 84-7*12 = 84-84 = 0
84 = 7*12+0
```
```
3) a=100 n=9   ->  11<100/9<12
r = a-qn = 100-11*(9) = 100-99 = 1
100 = 8*7+3
```
```
4) a=-96 n=12   ->  -8=-96/12
r = a-qn = -96-(-8)*(12) = -96+96 = 0
-96 = -8*12+0
```
